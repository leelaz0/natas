### krypton3

The password is encrypted under a substitution cipher to 
```
KSVVW BGSJD SVSIS VXBMN YQUUK BNWCU ANMJS
```
We begin our analysis by looking for known strings. Presumably, the string 'password' occurs in the plaintext. The only candidate location for this string is at `YQUUKBNW`, so we have a few substitutions.
```
wSVVdoGSJDSVSISVXoMrpasswordCsArMJS
```
We observe also that S occurs most frequently in the ciphertext and guess that it decrypts to 'e'
```
weVVdoGeJDeVeIeVXoMrpasswordCsArMJe
```
Look at the repeated letters in 'VeIeV', we guess that 'VeIeVXoMr' is 'levelfour'
```
welldoGeJDelevelfourpasswordCsAruJe
```
We can now guess that the first few characters are 'welldonethe'
```
welldonethelevelfourpasswordCsArute
```
And now plaintext is guessable
```
welldonethelevelfourpasswordisbrute
```

### krypton6

The password is encrypted with a stream cipher to 
```
PNUKLYLWRQKGKBE
```
We also have access to an encryption oracle. To begin analysing the cipher, we try encrypting a long repeated string
```
python3 -c "print('A'*200)" > /tmp/plaintext
/krypton/krypton6/encrypt6 /tmp/plaintext /tmp/ciphertext
cat /tmp/ciphertext
```
The output is as follows
```
EICTDGYIYZKTHNSIRFXYCPFUEOCKRN

EICTDGYIYZKTHNSIRFXYCPFUEOCKRN
EICTDGYIYZKTHNSIRFXYCPFUEOCKRN
EICTDGYIYZKTHNSIRFXY
```
We observe that the output is the same 30-character block repeated, thus the stream cipher is equivalent with a block cipher with the repeated block as key.

````python
a = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
k = 'EICTDGYIYZKTHNSIRFXYCPFUEOCKRN'
c = 'PNUKLYLWRQKGKBE'

def decrypt(c,k):
    r = ''
    for i in range(len(c)):
        n = a.index(k[i%len(k)])
        m = a.index(c[i])
        r += a[(m-n)%26]
    return r
    
print(decrypt(c,k))
````
